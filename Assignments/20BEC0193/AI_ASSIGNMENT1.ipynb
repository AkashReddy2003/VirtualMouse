<h1>Task1 - Create a pandas dataframe (DataFrame name as 'df') with numpy random values (4 features and 4 observation)</h1>

```python
import pandas as pd
import numpy as np
a=pd.Series(np.random.rand(4))
b=pd.Series(np.random.rand(4))
c=pd.Series(np.random.rand(4))
d=pd.Series(np.random.rand(4))

print(a,b,c,d)
```

```python
#Creating a database with the 4 series

df = pd.DataFrame([a,b,c,d])
print(df)
```

<h1>Task-2 : Rename the task - 1 'df' dataframe column names to 'Random value 1', 'Random value 2', 'Random value 3' & 'Random value 4'</h1>

```python
#Changing the column names

df.columns = ['Random value 1','Random value 2','Random value 3','Random value 4']
print(df)
```
